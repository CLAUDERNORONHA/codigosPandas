UTILIZANDO O APPLY PARA APLICAR FUNÃ‡OES EM DATAFRAME

```python
def consumo(a):
    if a <= 75:
        return 'baixo'
    elif a >= 75:
        return 'Alto'
    else:
        return 'Moderado'
```

```python
# cria uma coluna no Dataframe para saber o nivel de consumo de alcool
drinks['consumo'] = drinks['beer_servings'].apply(consumo) 
```

```python
drinks.head(20)
```

EXPORTANDO DATAFRAMES

```python
# sem indexaÃ§Ã£o
drinks.to_csv('drinks_consumo.csv', encoding='utf-8', index=False)
```

```python
# com indexaÃ§Ã£o
drinks.to_csv('drinks_indice.csv', encoding='utf-8', index=True)
```
