FILTRAR UM DATAFRAME

```python
# Vamos criar uma lista para saber as filmes com maior duração do que 180 mintos

boleanos = []
for item in filmes.duration:
    if item >=200:
        boleanos.append(True)
    else:
        boleanos.append(False)
```

```python
boleanos[:20]
```

```python
type(boleanos)
```

```python
# verificar o tamanho da lista
len(boleanos)
```

PRECISAOS CONVERTER A LISTA EM UM serie PARA APLICAR COM FILTRO DE DATAFRAME PANDAS

```python
#convertendo a lista e serie
duracao = pd.Series(boleanos)
duracao.head()
```
