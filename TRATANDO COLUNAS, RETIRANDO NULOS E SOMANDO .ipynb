#dropna()

DROPNA  = SE QUALQUER VALOR ESTIVER FALTANDO NA LINHA, ENTÃO O DROPNA APAGA A LINHA

```python
# apagar colunas nulas
B  = ufo.dropna(how='any').shape
```

```python
B
```

UTILIZANDO O DROPNA COM PARAMETRO


```python
# Estamos criando um subset com city e shaper.
# se houver qualquer valor nulo nessas colunas, a linha sera deletada

ufo.dropna(subset=['City','Shape Reported'], how='any').shape
```

```python
# Se todos os valores estiverem nulos em city ou Shape reported, as linhas deletadas
ufo.dropna(subset=['City','Shape Reported'], how='all').shape
```

```python
# PODEMOS AINDA PREENCHER OS VALORES QUE ESTÃO FALTANDO
ufo['Shape Reported'].fillna(value='DVERSOS', inplace=True)
```
